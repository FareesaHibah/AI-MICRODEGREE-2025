```python
import pandas as pd
import re
import string
import nltk
from nltk.corpus import stopwords
```

```python
df = pd.read_csv("sentiment-analysis-1.csv", )
df.head()
```

```python
# Function for converting the text to lower case

def lower_case(value):
    a = value.lower()
    return a
```

```python
# Function for removing punctuation

def remove_punc(value):
    b = value.translate(str.maketrans("", "", string.punctuation))
    return b
```

```python
# Function for removing digits and special characters

def remove_dig_char(value):
    c = re.sub(r'[^a-zA-Z\s]', '', value)
    return c
```

```python
# Function for removing white-spaces

def remove_spaces(value):
    d = re.sub(' +', ' ', value)
    return d
```

```python
# Function for removing stop-words

def remove_stp_words(value):
    e = nltk.word_tokenize(value)
    words = [i for i in e if i not in stopwords.words('english') ]
    return ' '.join(words)
```

```python
# Cleaning function
def clean(value):
    value = lower_case(value)
    value = remove_punc(value)
    value = remove_dig_char(value)
    value = remove_spaces(value)
    value = remove_stp_words(value)
    
    return value
```

```python
df = df.dropna()
df.isnull().sum
```

```python
df['Cleaned Text'] = df['Text'].apply(clean)
df.head()
```
