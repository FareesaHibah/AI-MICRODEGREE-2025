```python
# Data Visualization

df = pd.read_excel('Superstore.xlsx')
df
```

```python
category_sales = df.groupby('Category')['Sales'].sum()
category_sales
```

```python
plt.figure(figsize = (6, 4))
category_sales.plot(kind = 'line', marker = 'o', color = 'black')

# Add labels and titles
plt.xlabel("Category of Super store")
plt.ylabel("Total Sales")
plt.title("Total sales by each category")

# To add grid lines
plt.grid(True, linestyle = "--", alpha = 0.6)
```

```python
region_sales = df.groupby('Region')['Sales'].sum()
region_sales
```

```python
plt.figure(figsize = (6, 4))

# Bar chart
category_sales.plot(kind = 'bar', color = ['red', 'green', 'violet', 'orange'])

# Add labels and titles
plt.xlabel("Category of Super store")
plt.ylabel("Total Sales")
plt.title("Total sales by each category")

# To add grid lines
plt.grid(True, linestyle = "--", alpha = 0.6)
```
