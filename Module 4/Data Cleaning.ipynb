```python
# Imputation of missing data
# Filling the rows or column values with a particular statistical value(mean, median, mode and etc.)

''' mean = simple and effecient for your rows and columns values or dataset, does not consider relationship b/w 2 or more columns
median = less sensitive to outliers used for skew data

mode = best suitable to use with categorical data. We will fill with most frequent value'''

```

```python
x = df['Sales'].mean()
print('Mean:' ,x)
y = df['Sales'].median()
print('Median:', y)
```

```python
# Filling missing values:


# Filling missing values with mean: 
# Method 1:
df['Sales'] = df['Sales'].fillna(df['Sales'].mean())
df['Sales']

# Method 2:
df['Sales'] = df['Sales'].fillna(x)
df['Sales']


# Filling missing values with median: 
# Method 1:
df['Discount'] = df['Discount'].fillna(df['Sales'].median())
df['Discount']

# Method 2:
df['Discount'] = df['Discount'].fillna(y)
df['Discount']
```
